## TRAIN-TEST SPLIT

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X=df.drop("prognosis",axis=1)
Y=df["prognosis"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=20)
```

## TRAINING MODEL

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
```

```python
models={
    "SVC":SVC(kernel='linear'),
    "Random Forest":RandomForestClassifier(n_estimators=100,random_state=42),
    "Gradient Boosting":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "KNN":KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes":MultinomialNB()
}
for model_name, model in models.items():
    print(model_name,':',model)
```

```python
from re import A
for model_name, model in models.items():

    #train model
    model.fit(X_train,Y_train)

    #test model
    Y_pred=model.predict(X_test)

    #calculate accuracy
    accuracy=accuracy_score(Y_test,Y_pred)

    #calculate confusion matrix
    cm=confusion_matrix(Y_test,Y_pred)

    print(f"{model_name} Accuracy: {accuracy}")
    print(f"{model_name} Confusion Matrix: \n")
    print(np.array2string(cm,separator=','))

    print(model_name,accuracy_score(Y_test,Y_pred))
```

```python
symptoms=pd.read_csv("/content/drive/MyDrive/datasets/symtoms_df.csv")
precautions=pd.read_csv("/content/drive/MyDrive/datasets/precautions_df.csv")
workout=pd.read_csv("/content/drive/MyDrive/datasets/workout_df.csv")
description=pd.read_csv("/content/drive/MyDrive/datasets/description.csv")
medications=pd.read_csv("/content/drive/MyDrive/datasets/medications.csv")
diets=pd.read_csv("/content/drive/MyDrive/datasets/diets.csv")
```

```python
def helper(disease):
    # Description
    desc = description[description['Disease'] == disease]['Description']
    desc = ''.join(desc.values) if not desc.empty else "No description available."

    # Precautions
    pre = precautions[precautions['Disease'] == disease][['Precaution_1','Precaution_2','Precaution_3','Precaution_4']]
    pre = pre.values.flatten().tolist() if not pre.empty else ["No precautions listed."]

    # Medications
    med = medications[medications['Disease'] == disease]['Medication']
    med = med.values.flatten().tolist() if not med.empty else ["No medications listed."]

    # Diet
    diet = diets[diets['Disease'] == disease]['Diet']
    diet = diet.values.flatten().tolist() if not diet.empty else ["No dietary recommendations."]

    # Workout
    wrkout = workout[workout['disease'] == disease]['workout']
    wrkout = wrkout.values.flatten().tolist() if not wrkout.empty else ["No workout advice."]

    return desc, pre, med, diet, wrkout
```
